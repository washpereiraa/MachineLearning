```python
import pandas as pd
```

```python
# dicionario de paises 

paises1 = {
'Key': [1, 2 , 3],
'País': ['Brasil', 'Brasil', 'Brasil'],
'Estado': ['Minas Gerais', 'São Paulo', 'Rio de Janeiro'],
'População': [123465, 456789, 987654]}

paises2 = {
'Key': [2, 3 , 4],
'País': ['Estados Unidos', 'Argetina', 'Portugal'],
'Capital': ['Washington', 'Buenos Aires', 'Lisboa'],
'População': [223465, 356789, 487654]}

estados = {
'Key': [1, 2 , 3, 4],
'Estado': ['Minas Gerais', 'São Paulo', 'Rio de Janeiro', 'Amazonas'],
'Capital': ['Belo Horizonte', 'Cidade de São Paulo', 'Cidade do Rio de Janeiro', 'Manaus'],
'População': [122465, 135678, 147654, 6548]}


dfPaises1 = pd.DataFrame(paises1, columns=['Key','País','Estado','População'])
dfPaises2 = pd.DataFrame(paises2, columns=['Key','País','Capital','População'])
estados = pd.DataFrame(estados, columns=['Key','Estado','Capital','População'])
```

```python
dfPaises1
```

```python
estados
```

```python
#Metodo CONCAT = coloca todos dados juntos sem nunhum filtro
# por padrão os dataframes sao cacatenados pelo eixo 0/linhas = axis = 0

pd.concat([dfPaises1, dfPaises2])
```

```python
#Metodo CONCAT = coloca todos dados juntos sem nunhum filtro
# concatenando por colunas 

pd.concat([dfPaises1, dfPaises2], axis = 1)
```

```python
#Metodo MERGE
#inner = mescla onde as chaves forem iguais

pd.merge(dfPaises1, estados, how='inner', on=['Key'])
```

```python
#Metodo MERGE
#right/left = atribui os dados coforme passados as direções

pd.merge(dfPaises1, estados, left_on='Key', right_on='Key')
```

```python
#Metodo JOIN

df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],
                   'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],
                      'B': ['B0', 'B1', 'B2']})

df.join(other, lsuffix='_caller', rsuffix='_other')
```

```python
dfPaises1.join(estados, lsuffix='_caller')
```
