```python
#importar Cufflinks

import cufflinks as cf
cf.go_offline
```

```python
#O plotly e o cufflinks nao tem banco de dados nativos para trabalhos experimentais, então sera necessario cria-los no pandas
# e numpy para sevir como objeto de estudo

dfAleatorio = pd.DataFrame(np.random.randn(100, 4), columns=['A', 'B', 'C', 'D'])
dfAleatorio.head()
```

```python
dfAleatorio2 = pd.DataFrame({'Categoria':['A', 'B', 'C'], 'Valores':[10, 20, 30]})
dfAleatorio2
```

```python
#Plotando um grafico usando Plotly - iplot

dados = go.Scatter(x = dfAleatorio['A'], y = dfAleatorio['B'], mode = 'markers')
grafico = [dados]
py.iplot(grafico)
```

```python
#exemplo simples

iplot([{'x': [1, 2, 3], 'y': [5, 2, 7]}])
```

```python
trace = go.Box(x=dfAleatorio2['Categoria'], y=dfAleatorio2['Valores'])
grafico = [trace]
py.iplot(grafico)
```

```python
DfTridimencional = pd.DataFrame({'x':[1,2,3,4], 'y':[4,3,2,1], 'z':[1,2,3,4]})

trace = go.Surface(x=DfTridimencional['x'], y=DfTridimencional['y'], z=DfTridimencional['z'])
grafico = [trace]
py.update_layout(title_text="Ring cyclide")
```
