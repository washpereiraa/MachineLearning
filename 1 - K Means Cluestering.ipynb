```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

#criar um dataset aleatorio para teste
from sklearn.datasets import make_blobs

#importando K-Means
from sklearn.cluster import KMeans
```

```python
dataMB = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std = 1.8, random_state = 50)
```

```python
type(dataMB)
```

```python
#visualizando 

plt.figure(figsize = [12, 8])

#dados do grafico:
                #X               #Y               #clases
plt.scatter(dataMB[0][:, 0], dataMB[0][:, 1], c = dataMB[1], cmap='rainbow')]

#numero de cluster determinado pela quantidade de cores
```

```python
#estanciando objeto KMeans

kmeans = KMeans(n_clusters = 4)


#numero de cluster determinado pela quantidade de cores
```

```python
#fit do modelo

kmeans.fit(dataMB[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
#criando grafico

figure, (grafico1, grafico2) = plt.subplots(1 , 2, figsize = (12, 8), sharey = True)

#primeiro grafico
grafico1.set_title('kMeans')
grafico1.scatter(dataMB[0][:, 0], dataMB[0][:, 1], c = kmeans.labels_, cmap='rainbow')

#segundo grafico
grafico2.set_title('Original')
grafico2.scatter(dataMB[0][:, 0], dataMB[0][:, 1], c = dataMB[1], cmap='rainbow')
```
