```python
#COMEÇANDO O ALGORITMO DE REGRESSÃO LOGISTICA

from sklearn.model_selection import train_test_split as trainTest
```

```python
#divisão dos dados
#X = Tudo que iremos utilizar para prever o Clique no anuncio 
#Y = o data a ser previsto 

x = df_adData[['Daily Time Spent on Site', 'Age', 'Area Income', 'Daily Internet Usage', 'Male']]
y = df_adData['Clicked on Ad']
```

```python
#triando os dados

xTrain, xTest, yTrain, yTest = trainTest(x, y, test_size=0.33, random_state=42) 
#radom_state = 42 para manter a mesma resposta que o professor
```

```python
#Ajustando modelo de treinamento
from sklearn.linear_model import LinearRegression

logModel = LinearRegression()

logModel.fit(xTrain, yTrain)
```

```python
#Previsão e avaliação
#Previsaõ feita com os dados de teste

predict = logModel.predict(xTest)
```

```python
#Relatorio do Modelo

from sklearn.metrics import classification_report
```

```python
#Comparando os dados de teste X Acertos
print(classification_report(yTest, predict))
```
