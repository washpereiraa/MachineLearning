```python
import numpy as np
vetorx = np.arange(0,100)
vetory = vetorx * 2
vetorz = vetorx **2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#caso nao acrescente a %matplotlib inline, tem que fazer o uso de plt.show()
```

```python
#Estancia um objeto grafico, plotar um grafico, informar a posição e tiulo do grafico

objetoGrafico = plt.figure() #estanciando objeto
grafico = objetoGrafico.add_axes([0,0,1,1]) #definindo posição
grafico.plot(vetorx, vetory) #plotando grafico
grafico.set_title('Title') #inserindo um titulo
grafico.set_xlabel('EIXO X')
grafico.set_ylabel('EIXO Y')
```

```python
#Crie dois graficos, um dentro/sobre do outro, definindo tamanho e posição

graficoAX = plt.figure()

graficoAX1 = graficoAX.add_axes([0,0,1,1]) #posição e tamnho do grafico
graficoAX1.plot(vetorx, vetory) #plotando o grafico

subGrafico = graficoAX.add_axes([0.1,0.6,0.3,0.3]) #posição e tamnho do subgrafico
subGrafico.plot(vetorx, vetory)
```

```python
#Criar dois graficos com eixos X, Y e Z , no segundo grafico selecionar um area especifica para visualização

objGrafico = plt.figure()

graficoAXES1 = objGrafico.add_axes([0,0,1,1])
graficoAXES1.plot(vetorx, vetorz)
graficoAXES1.set_title('Grafico') #inserindo um titulo
graficoAXES1.set_xlabel('EIXO X')
graficoAXES1.set_ylabel('EIXO Z')

subGrafico01 = objGrafico.add_axes([0.1,0.6,0.3,0.3])
subGrafico01.plot(vetorx, vetory)
subGrafico01.set_title('ZOOM') #inserindo um titulo
subGrafico01.set_xlabel('EIXO X')
subGrafico01.set_ylabel('EIXO Y')
#selecionando area de exibição do grafico 
subGrafico01.set_xlim(20,22)
subGrafico01.set_ylim(30,50)
```

```python
#criar dois grafico através do metodo subplots

fig, axes = plt.subplots(nrows= 1, ncols= 2)
```

```python
#plotar dois graficos acima através dos métodos de subplots

axes[0].plot(vetorx, vetory, color='blue', ls='--')
axes[1].plot(vetorx, vetorz, color='red', ls='-')

fig
```

```python
#redimencionar os graficos acima

fig, axes = plt.subplots(nrows= 1, ncols= 2, figsize=(12,3))
axes[0].plot(vetorx, vetory, color='blue', ls='--')
axes[1].plot(vetorx, vetorz, color='red', ls='-')
```
