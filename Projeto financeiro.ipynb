```python
from pandas_datareader import data, wb
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#o datareader exige 4 parametro: plataforma onde esta a base de dados, base de dados, inicio do perido de analise, fim do
#periodo de analise

#Variaveis
# BAC = Banco da America
# inicio = data inicial
# fim = data final

inicio = datetime.datetime(2006, 1 , 1)
fim = datetime.datetime(2016, 1 , 1)

dfBAC = data.DataReader("BAC", "yahoo", inicio, fim)

dfBAC
```

```python
#CRie um grafico linear simples das colunas OPEN E CLOSE

dfBAC["Open"].plot()
```

```python
dfBAC['Close'].plot()
```

```python
#importando outras base de dados

#CitiGroup
dfC = data.DataReader("C", "yahoo", inicio, fim)

#Golden Sachs
dfGS = data.DataReader("GS", "yahoo", inicio, fim)

#JPMorgan Chase
dfJPM = data.DataReader("JPM", "yahoo", inicio, fim)

#Morgan Stanley
dfMS = data.DataReader("MS", "yahoo", inicio, fim)

#Wells Fargo
dfWFC = data.DataReader("WFC", "yahoo", inicio, fim)
```

```python
#Crie uma lista com o nome dos bancos e chame-a de ListadeBancos

ListadeBancos = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
#Concatene todos os dataframes e um novo chamdo TodosBancos e use a lista tickers para nomear as colunas

dfTodosBancos = pd.concat([dfBAC, dfC, dfGS, dfJPM, dfMS, dfWFC], axis= 1, keys= ListadeBancos)
```

```python
#Definindo niveis de colunas

dfTodosBancos.columns.names =['Nome dos Bancos', 'infos']

dfTodosBancos
```

```python
#Descubra o valor maximo de fechamento de cada banco

dfTodosBancos.xs(key='Close', axis=1, level='infos').max()
```

```python
#Crie um data frame chamdo dfRetorno, esse dataframe contera as informaÃ§Ãµes de ativos percentuais de cada banco
#formular para descobrir os ativos percentuais

# retorno = contaÃ§Ã£o do dia - contaÃ§Ã£o do dia anterior / contaÃ§Ã£o do dia

dfRetorno = pd.DataFrame()
```

```python
#continua

for nomedoBanco in ListadeBancos:
    dfRetorno[nomedoBanco + 'Retorno'] = dfTodosBancos[nomedoBanco]['Close'].pct_change()
```
