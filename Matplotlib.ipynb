```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
vetorx = np.linspace(0, 5, 11)
vetory = vetorx * vetorx
```

```python
#Metodo PLOT

plt.plot(vetorx, vetory)
plt.xlabel('EIXO X')
plt.ylabel('EIXO Y')
plt.title('PRIMEIRO GRAFICO')
```

```python
#Metodo SUBPLOT
#primeiro parametro = quantas linhas de graficos
#segundo = quantas colunas de graficos
#terceiro = seleção de qual grafico esta trabalhando segundo os parametros acima

plt.subplot(1, 2, 1)
plt.plot(vetorx, vetory)

plt.subplot(1, 2, 2)
plt.plot(vetory, vetorx, color='red')
```

```python
#Melhor forma de trabalhar com graficos em Matplotlib é estanciar objetos para ter maior controle de criação

#estanciando objeto
objGrafico = plt.figure()

#craindo um grafico personalizado
grafico = objGrafico.add_axes([0.1, 0.1, 0.8, 0.8])
grafico.plot(vetorx, vetory, color='red')
grafico.set_xlabel('EIXO X')
grafico.set_ylabel('EIXO Y')
grafico.set_title('SEGUNDO GRAFICO')

#criando um segundo grafico com informações complementares ao primeiro grafico
subGrafico = objGrafico.add_axes([0.2, 0.5, 0.3, 0.3])
subGrafico.plot(vetory, vetorx, color='blue')
subGrafico.set_xlabel('EIXO X')
subGrafico.set_ylabel('EIXO Y')
subGrafico.set_title('SEGUNDO GRAFICO')
```

```python
#SUBPLOT - ESTACIAMENTO/OBJETO

grafico, eixos = plt.subplots()
eixos.plot(vetorx, vetory ** 3, 'b--', label='dados 1', linewidth=3, alpha=0.5)
eixos.plot(vetorx, vetory ** 2, 'r--', label='dados 2')
eixos.set_xlabel('eixo x')
eixos.set_ylabel('eixo y')
eixos.legend()
```

```python
#SUBPLOT - ESTACIAMENTO/OBJETO - em lista

grafico, eixos = plt.subplots(nrows= 1, ncols=2, figsize=(10,3))

eixos[0].plot(vetorx, vetory ** 3, 'b--')
eixos[0].set_xlabel('eixo x')

eixos[1].plot(vetorx, vetory ** 2, 'r--')
eixos[1].set_ylabel('eixo y')
```

```python
#SUBPLOT - ESTACIAMENTO/OBJETO - em lista

for axe in eixos:
    axe.plot(vetorx, vetory, 'r--' )
    axe.plot(vetorx, vetory, 'b--')
grafico
```
