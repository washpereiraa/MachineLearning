```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#Exercios feito na base de datos do Titanic

titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#1 Jointplot entre as colunas FARE e AGE, sendo a primeira o eixo X

sns.jointplot(x='fare', y='age', data=titanic)
```

```python
#2 Distplot

sns.distplot(titanic['fare'], kde=False, color='red', bins=30)
```

```python
#3 boxplot

sns.boxplot(x='class', y='age', data=titanic)
```

```python
#4 Swarmplot

sns.swarmplot(y='age', x='class', data=titanic, palette='Set2')
```

```python
#5 Countplot - grafico de idade

sns.countplot('sex', data=titanic)
```

```python
#6 Correlação e mapa de calor

corr = titanic.corr()
sns.heatmap(corr, cmap='coolwarm')
plt.title('Correlação Titanic')
```

```python
#7 FaceGrid

fgTitanic = sns.FacetGrid(titanic, col='sex')
fgTitanic.map(plt.hist, 'age')
```
